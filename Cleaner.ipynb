```python
import numpy as np
import pandas as pd

pd.options.mode.chained_assignment = None
```

```python
#import the scraped dataset
df = pd.read_csv('marathon_raw.csv', sep=',')
df.head(5)
```

First, we adjust the variables and transform them to the format convinient for the further analysis. Transformations are commented in each code chunk.

```python
df.info()
```

```python
#from the first column get the country abbreviation
df['Country'] = df['Name'].str[-4:-1]
df.head(5)
```

```python
#remove the country from the name column - last 6 characters
df['Name'] = df['Name'].str[:-6]
df.head(5)
```

```python
#get the gender from the category (Age_group)
df['Gender'] = df['Age_group'].str[0:1]
df.head(5)
```

```python
#similarly, get the age category from the same column (we still want to preserve the original column, might be useful later)
df['Age'] = df['Age_group'].str[1:]
df.head(5)
```

```python
#in the age column, we got 2 non-numerical values
#H should refer to the 20-29 age category, for consistency, replace with 20
#similarly, Jug should refer to 18-20, replace with 18
df["Age"].replace({"H": "20", "Jug": "18"}, inplace=True)
df.head(5)
```

```python
#in the analysis, we want to refer to the countries by its full name, upload the csv we have downloaded
countries = pd.read_csv('country-codes.csv')
countries.head(5)
```

```python
#get more friendly column names (of those that we need) - IOC code is used in the marathon dataset 
countries.rename(columns={'IOC': 'Country', 'CLDR display name': 'Country_long'}, inplace=True)
#keep only those two columns for easier manipulation
countries = countries[["Country", "Country_long"]]
```

```python
#when merging (compare resulting rows of inner and left join) -> several countries have different codes
#case of Taiwan (Taipei), Montenegro, Trinidad and Tobago and Faroe Islands
#Kosovo is not there at all - probably unrecognised by the time the dataset was created - we add it
countries["Country"].replace({"TPE":"TWN", "MGO":"MNE", "TTO":"TRI", "FAR":"FRO"}, inplace = True)
countries.loc[countries.shape[0]] = ['KOS', 'Kosovo']
```

```python
#and merge it to the runners - only we need only the full name of the country
results = pd.merge(df, 
                      countries, 
                      on ='Country', 
                      how ='inner')
```

```python
results
```

```python
#double check, that every runner's country got the full name assigned by inner join
df.shape[0] == results.shape[0]
```

```python
#transform Time, First_half and Second_half to minutes; Pace_0_5 and Pace_35_40 to seconds
results["Time_mins"] = results["Time"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]) + int(x[2])/60).round(1)

#code above works nicely, but the split times (paces) are sometimes missing and the dash causes an error - replace it
results["First_half"].replace({'–':'00:00:00'}, inplace = True)
results["Second_half"].replace({'–':'00:00:00'}, inplace = True)
results["Pace_0_5"].replace({'–':'00:00'}, inplace = True)
results["Pace_5_10"].replace({'–':'00:00'}, inplace = True)
results["Pace_10_15"].replace({'–':'00:00'}, inplace = True)
results["Pace_15_20"].replace({'–':'00:00'}, inplace = True)
results["Pace_20_21"].replace({'–':'00:00'}, inplace = True)
results["Pace_21_25"].replace({'–':'00:00'}, inplace = True)
results["Pace_25_30"].replace({'–':'00:00'}, inplace = True)
results["Pace_30_35"].replace({'–':'00:00'}, inplace = True)
results["Pace_35_40"].replace({'–':'00:00'}, inplace = True)
results["Pace_40_42"].replace({'–':'00:00'}, inplace = True)

#this way we dot not entirely get rid of those with missing splits, we can filter them out in each analysis
results["First_half_mins"] = results["First_half"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]) + int(x[2])/60).round(1)
results["Second_half_mins"] = results["Second_half"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]) + int(x[2])/60).round(1)
results["Pace_0_5_sec"] = results["Pace_0_5"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_5_10_sec"] = results["Pace_5_10"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_10_15_sec"] = results["Pace_10_15"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_15_20_sec"] = results["Pace_15_20"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_20_21_sec"] = results["Pace_20_21"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_21_25_sec"] = results["Pace_21_25"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_25_30_sec"] = results["Pace_25_30"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_30_35_sec"] = results["Pace_30_35"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_35_40_sec"] = results["Pace_35_40"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
results["Pace_40_42_sec"] = results["Pace_40_42"].str.split(':').apply(lambda x:int(x[0])*60 + int(x[1]))
```

```python
#write to the new csv
results.to_csv('marathon_clean.csv', encoding='utf-8', index=False)
```

We create the second dataset where we keep only the pace columns (and names) which will be used to graph the paces in the plot for each runner separately.

```python
#we select only the the name and pace rows (end with _sec)
pace_table = results.filter(regex='sec$|Name$', axis=1)

#we remove accents from all the names, otherwise it creates some inconsistencies when sorting (e.g. á goes after z)
pace_table['Name'] = pace_table['Name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

#transform the table to the long format
pace_table = pd.melt(pace_table, id_vars= pace_table.columns[0], value_vars= pace_table.columns[1:])


#the pace will be plotted on the y-axis, the kilometer will be shown on the x axis 
#transform the initial variable name, it is the third element of the name (which is segmented by underscore)
pace_table['variable'] = pace_table['variable'].str.split('_').str[2]
#and we rename the columns
pace_table = pace_table.rename({'variable': 'Km', 'value': 'Pace'}, axis=1)
#and set the kilometers to numeric
pace_table['Km'] = pd.to_numeric(pace_table['Km'])
#finally, we sort the table by the name, as it will be used in the drop down menu later
pace_table = pace_table.sort_values('Name')
```

```python
#write to the csv
pace_table.to_csv('pace_table.csv', encoding='utf-8', index=False)
```
